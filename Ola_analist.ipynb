```python
data['datetime']=pd.to_datetime(data['datetime'],format="%d/%m/%Y")
data.info()
```

```python
data.head(10)
```

```python
data["years"]=data["datetime"].dt.year
data ["months"]=data["datetime"].dt.month
data.head ()
```

```python

total_bulan_tahun = data.groupby(['years', 'months'])['count'].sum()
print(total_bulan_tahun)


```

```python
total_bulan_tahun = data.groupby(['years', 'months'])['count'].sum().reset_index()
```

```python
import matplotlib.pyplot as plt

# gabungkan label year-month
total_bulan_tahun['label'] = total_bulan_tahun['years'].astype(str) + '-' + total_bulan_tahun['months'].astype(str)

plt.figure(figsize=(12,5))
plt.bar(total_bulan_tahun['label'], total_bulan_tahun['count'])
plt.xticks(rotation=45)
plt.xlabel("Year-Month")
plt.ylabel("Total Count")
plt.title("Total Count per Month per Year")
plt.tight_layout()
plt.show()
```

```python
# dataf register vs casual
label=["casual", "registered"]
sum_values=[data["casual"].sum(),data["registered"].sum()]

plt.pie(sum_values,labels=label)
plt.title ("casual vs registed orders")
plt.show()
```

```python
from collections import Counter
labels=[]
for label in data["season"]:
    if label == 1 :
        lab_="spring"
    elif label==2:
        lab_="summer"
    elif label==3:
        lab_="fall"
    else:
        lab_="winter"
    labels.append(lab_)

labels_val=Counter(labels)
assert len(labels_val.values())==4
assert len(labels_val.keys())==4
labels_val
```

```python
keys=labels_val.keys()
values=labels_val.values()
plt.pie (values,labels=keys)
plt.title("season")
plt.show ()
```

```python
label_weathers=[]
for label in data["weather"]:
    if label==1:
        labs_="sunny"
    elif label==2:
        labs_="cloudy"
    elif label==3:
        labs_="snow"
    else: 
        labs_="strom"
    label_weathers.append(labs_)

labels_wheaters=Counter(label_weathers)
```

```python
keys_weather=labels_wheaters.keys()
values_weather=labels_wheaters.values()

plt.pie(values_weather,labels=keys_weather)
plt.title("weather")
plt.show()
```
